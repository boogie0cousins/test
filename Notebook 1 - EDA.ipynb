Below are some utility functions to assist in plotting different charts.

```python
def get_color(column):
    """
    Get color for plot depending upon column name
    """
    return 'lightcoral' if column == 'Signal' else 'dodgerblue'

def get_colors(column):
    """
    Get pair of colors depending upon column name
    """
    return ('lightcoral', 'maroon') if column == 'Signal' else ('dodgerblue', 'blue')

    
def scatter_plot(x, y, ax):
    """
    Scatter plot between signal and close price
    """
    ax.scatter(x=x, y=y, marker='o', color='sandybrown', edgecolor='black')
    ax.set(title='Scatter Plot of Signal vs ClosePrice',
           xlabel='Signal',
           ylabel='ClosePrice')

def add_legend(ax):
    """
    Add legend to given axes
    """
    leg = ax.legend(loc='best', frameon=True, framealpha=1)
    leg.get_frame().set_edgecolor('black')
    
    
def cross_correlation_plot(df, x, y, ax):
    """
    Cross-correlation plot
    """
    xseries = df[x]
    yseries = df[y]
    
    # Calculates lags and correlation values
    lags, c, line, b = ax.xcorr(xseries, yseries, usevlines=True, maxlags=50, normed=True, lw=2)
    
    # Identify lag with max correlation value
    # and highlight that point
    max_val_idx = c.argmax()
    ax.axvline(x=lags[max_val_idx], color='lime', linestyle='-', 
               label=f'Max correlation = {c[max_val_idx].round(2)} at Lag = {lags[max_val_idx]} of {x}')
    
    ax.set(title='Cross-correlation between Signal and ClosePrice',
           xlabel='Lags',
           ylabel='Pearson correlation')
    
    add_legend(ax)
    
    
def rolling_correlation_plot(df, ax, sma_days=[10, 21]):
    """
    Rolling correlation plot
    """
    for days in sma_days:
        rolling_corr_df = df['Signal'].rolling(window=days, center=True).corr(df['ClosePrice'])
        rolling_corr_df.plot(ax=ax, label=f'{days}-Day Rolling Pearson Correlation')
        
    ax.set(title='Rolling Pearson Correlation between Signal and ClosePrice [{0}]'.format(", ".join(['{0} days'.format(str(x)) for x in sma_days])),
           xlabel='Date',
           ylabel='Correlation')
    add_legend(ax)
    

def line_plot(df, column, ax, sma_days=10):
    """
    Line plot for given time series
    """
    color1, color2 = get_colors(column)
    series = df[column]
    series.plot(ax=ax, color=color1, label=f'{column}')
    series.rolling(sma_days).mean().plot(ax=ax, color=color2, alpha=0.7, label=f'{column} {sma_days}-day Moving Average')
    ax.set(title=f'{column} time series',
           xlabel='Date',
           ylabel=f'{column}')
    add_legend(ax)
    

def distribution_plot(df, column, ax):
    """
    Distribution plot
    """
    color = get_color(column)
    series = df[column]
    sns.distplot(series, color=color, ax=ax, 
                 hist_kws=dict(edgecolor="k", linewidth=2), 
                 bins=30, hist=True, label=f'{column}')
    
    ax.set_title(f'Distribution of {column}')
    add_legend(ax)
    

def box_plot(df, column, ax):
    """
    Box plot
    """
    series = df[column]
    color = get_color(column)
    sns.boxplot(series, color=color, ax=ax)
    ax.set(title=f'Box plot for {column}')
```

```python
def plots_for_outliers(df):
    """
    Useful plots for checking outliers time-series data
    """
    signal = df['Signal']
    close_price = df['ClosePrice']
    
    grid = gridspec.GridSpec(nrows=5, ncols=2)

    fig = plt.figure(figsize=(15, 15))
    fig.clf()

    # Add axes which can span multiple grid boxes
    ax1 = fig.add_subplot(grid[0, :])
    ax2 = fig.add_subplot(grid[1, 0])
    ax3 = fig.add_subplot(grid[1, 1])
    ax4 = fig.add_subplot(grid[2, 0])
    ax5 = fig.add_subplot(grid[2, 1])
    ax6 = fig.add_subplot(grid[3, 0])
    ax7 = fig.add_subplot(grid[3, 1])
    
    # Scatter plot
    scatter_plot(x=signal, y=close_price, ax=ax1)
    
    # Time series plots
    line_plot(df, column='Signal', ax=ax2, sma_days=10)
    line_plot(df, column='ClosePrice', ax=ax3, sma_days=10)
    
    # Distribution plots
    distribution_plot(df, column='Signal', ax=ax4)
    distribution_plot(df, column='ClosePrice', ax=ax5)
        
    # Box plots
    box_plot(df, column='Signal', ax=ax6)
    box_plot(df, column='ClosePrice', ax=ax7)
    
    fig.tight_layout()
    plt.show()
```

## Identify Outliers

There are multiple ways to identify outliers given a dataset. Few of the these include:

1. Z-score
2. Modified Z-score
3. Interquartile range

```python
def get_numpy_array(series):
    """
    Convert a series to numpy.array
    """
    if not isinstance(series, np.ndarray):
        series = np.array(series)
    return series

def get_zscore(series):
    """
    Get Z-score using scipy.stats method
    """
    return np.abs(stats.zscore(series))


def get_outliers_using_zscore(series, threshold=3):
    """
    Identify the outliers using Z-score method
    """
    series = get_numpy_array(series)    
    z_score = get_zscore(series)
    return series[np.where(z_score > threshold)]


def get_outliers_using_modified_zscore(series, threshold=3.5):
    """
    Identify the outliers using modified Z-score method. 
    Overcome the limitations of Z-score method.
    """
    series = get_numpy_array(series)
    median = np.median(series)
    median_absolute_deviation = np.median([np.abs(val - median) for val in series])
    modified_z_scores = [0.6745 * (val - median) / median_absolute_deviation for val in series]
    return series[np.where(np.abs(modified_z_scores) > threshold)]


def get_iqr_range(series):
    """
    Get Interquartile range for given series
    """
    series = get_numpy_array(series)
    q1, q3 = np.percentile(series, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    return lower_bound, upper_bound
    

def get_outliers_using_iqr(series):
    """
    Identify the outliers using interquartile range (IQR)
    """
    series = get_numpy_array(series)
    lbound, ubound = get_iqr_range(series)
    return [x for x in series if x < lbound or x > ubound]
```

```python
def print_outliers(method, column):
    """
    Print outliers identified using given method and for given column of data.
    """
    if method == 'zscore':
        outliers = get_outliers_using_zscore(data[column])
    elif method == 'modified_zscore':
        outliers = get_outliers_using_modified_zscore(data[column])
    elif method == 'iqr':
        outliers = get_outliers_using_iqr(data[column])
    else:
        raise ValueError("Incorrect method value. Valid values: [zscore, modified_zscore, iqr]")
    
    print("\n{0} Outliers using '{1}' method: ".format(column, method), *outliers, sep="\n => ")
```

## Fix the outliers using IQR method

```python
# Make a copy of the data
clean_data = data.copy()

for col in clean_data.columns:
    # Get lower and upper bound for IQR and set outliers to NaN
    lbound, ubound = get_iqr_range(clean_data[col])
    clean_data[col] = np.where(np.logical_or(clean_data[col] < lbound, clean_data[col] > ubound), np.nan, clean_data[col])
    
    # Interpolate missing values
    clean_data[col] = clean_data[col].interpolate()
```

# Modeling

Since there exists a linear relationship between `Signal` and `ClosePrice` after cleaning the data (i.e. from the earlier plots), let's apply simple linear regression to see if our independent variable `Signal` helps us to predict dependent variable `ClosePrice` of SPY.

```python
def linear_regression(df, y_col, x_col=None, plot=False):
    """
    Perform linear regression and plot the regression line.
    """
    
    x_cols = []
    if x_col is None: 
        # If multiple regression, add all columns except y as X-variables
        x_cols = [col for col in df.columns if col != y_col]
    else:
        # Simple linear regression with just 1 independent variable
        x_cols = [x_col]

    X = df[x_cols]
    # adding a intercept
    X = sm.add_constant(X) 
    Y = df[y_col]

    model = sm.OLS(Y, X).fit()
    print_model = model.summary()
    print(print_model)
    
    # Line of best fit
    if plot:
        fig, ax = plt.subplots(figsize=(15, 10))
        clean_data.plot(x=x_col, y=y_col, kind='scatter', ax=ax, label='Data Points')
        abline_plot(model_results=model, ax=ax, label='Fitted line')
        ax.set(title='Line of best-fit for Signal vs ClosePrice',
               xlabel=x_col, 
               ylabel=y_col)
        add_legend(ax)

        plt.show()

    return model
```

```python
model = linear_regression(clean_data, x_col='Signal', y_col='ClosePrice', plot=True)
```

# Quality Check of Model

Quality check of our fitted model is very important to understand if the fitted model does not violate the assumptions behind this simple linear model. 

These assumptions are:

1. Gauss-Markov Theorem: OLS estimator gives the best linear unbiased estimate (BLUE) of the coefficients, given:
    - Mean or Expectation of error (residuals) is 0
    - Errors are uncorrelated
    - Errors are homoscedastic (i.e. they have equal variance)
2. Independent variables and residuals are uncorrelated
3. Normality of residuals 


Let's verify if these assumptions are not violated one-by-one.

## Mean of residuals should be 0

```python
np.mean(model.resid)
```

> <font color='red'>**Comment:**
> - Mean of residual is very small (approximately equal to 0). So this assumption is not violated.

## Residuals are homoscedastic

To check if residuals are homoscedastic, we can look at the plot of residuals vs predicted (fitted) values and perform Breusch-Pagan test.

### Residual vs Fitted values plot

```python
def homoscedastic_plot(model):
    """
    Plot for homoscedasticity check
    """
    fitted_values = model.predict()
    residuals = model.resid
    residuals_standardized = model.get_influence().resid_studentized_internal
    
    fig, ax = plt.subplots(1, 2, figsize=(15, 7))
    
    sns.regplot(x=fitted_values, y=residuals, lowess=True, ax=ax[0], line_kws={'color': 'red'})
    ax[0].set(title='Residuals vs Fitted', 
              xlabel='Fitted Values',
              ylabel='Residuals')
    
    sns.regplot(x=fitted_values, y=np.sqrt(np.abs(residuals_standardized)), lowess=True, ax=ax[1], line_kws={'color': 'red'})
    ax[1].set(title='Scale-Location', 
              xlabel='Fitted Values',
              ylabel='sqrt[abs(residuals)]')
    
    fig.tight_layout()
    plt.show()
```

```python
homoscedastic_plot(model)
```

> <font color='red'>**Comment:**
> - The points in the above plots should be random and no peculiar (increase or decrease) pattern in the values of residuals should be visible. Red line is relatively flat in our Residual vs Fitted plot. This indicates the errors are homoscedastic.

### Breusch-Pagan Test

```python
bp = pd.DataFrame(columns=['Values'], index=['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value'])
bp['Values'] = sms.het_breuschpagan(model.resid, model.model.exog)
bp
```

> <font color='red'>**Comment:**
> - p-value in LM and F-test is very low (< 0.05). Thus, we reject the null hypothesis that errors (or residuals) are homoscedastic. Point to note that this test can give low p-value when the residuals are not heteroscedastic if the model is flawed (may be due to ommited variable bias). 
>                                                 
> - Given the line in residual vs fitted value plot is quite flat, we can conclude that residuals are homoscedastic and Breusch-Pagan test is rejecting the null hypothesis due to some other flaws in the model.

## Residuals are uncorrelated 

Residuals should be uncorrelated. We can test it by plotting ACF (auto correlation function) plot and Durbin-Watson test.

### Plotting Autocorrelation Function

```python
def plot_acf(model):
    """
    Plot autocorrelation function
    """
    fig, ax = plt.subplots(figsize=(15, 7))    
    acf = smt.graphics.plot_acf(model.resid, lags=50, alpha=0.05, ax=ax)
    acf.show()
```

```python
plot_acf(model)
```

### Durbin-Watson Test

This test statistic is used to detect the presence of autocorrelation at lag 1 in the residuals. Value of the test statistic is always between 0 and 4. If the value is 2, then there is no autocorrelation. If value < 2, then positive autocorrelation in residuals. If value > 2, negative autocorrelation in residuals exists.

> <font color='red'>**Comment:**

### DateTime-based features

```python
def add_datetime_features(df):
    """
    Add features based on datetime index.
    """
    _df = df.copy()
    
    if not isinstance(_df.index, pd.DatetimeIndex):
        _df.index = pd.to_datetime(_df.index)
        
    # Add year, quarter, month, week, day
    _df['Year'] = _df.index.year
    _df['Quarter'] = _df.index.quarter
    _df['Month'] = _df.index.month
    _df['Week'] = _df.index.week
    _df['Day'] = _df.index.day
    
    # Add day of week
    _df['DayofWeek'] = _df.index.dayofweek

    return _df
```
